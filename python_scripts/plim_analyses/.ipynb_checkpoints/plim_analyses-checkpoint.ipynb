```python
import os, sys
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.signal import find_peaks
```

```python
data_dir = "../../data/2022-11-22"
```

```python
dirnames = os.listdir(data_dir)
dirnames = list(os.path.join(data_dir, f) for f in dirnames)

fnames = []

for dirname in dirnames:
    files = os.listdir(dirname)
    for file in files:
        if ".dat" in file:
            fnames.append(f"{dirname}/{file}")
    break
```

```python
data = []
for fname in fnames:
    df = pd.read_csv(fname)
    df["t"] = df["T_(ms)"]/1000 #ms to s
    df["y"] = 90.08745-((df["V_dist_(mA)"]-4)/16)*101.60 #convert milliamp to millimeters
    df["p_c"] = (df["Comp_P_(mA)"]-4)*0.75
    df["v"] = df["Sys_Volts_(V)"]
    df["l"] = df["LC_(grams)"]
    df["p_p"] = (df["Piezo_P_(mA)"]-4)*0.75
    df["p_o"] = 6*(df["Piezo_out_(mV)"]-4000)/16000 #converting milliamp output to expected bar output
    df.style.set_caption(fname.split('/')[-1].split('.')[0])
    data.append(df)
```

```python
fig, axes = plt.subplots(5,1, sharex=True, figsize=(7,7))
axes[0].plot(data[0]["t"], data[0]["y"],linestyle='None',marker='o',markersize=1)
axes[0].grid()
axes[0].set_ylabel("Distance (mm)")

axes[1].plot(data[0]["t"], data[0]["p_o"],linestyle='None',marker='o',markersize=1)
axes[1].grid()
axes[1].plot(data[0]["t"], data[0]["p_p"],linestyle='None',marker='o',markersize=1)
axes[1].set_ylabel("Pressure - Actuator \n(Bar)")
axes[1].set_ylim([-0.25, 6.25])

axes[2].plot(data[0]["t"], data[0]["p_c"],linestyle='None',marker='o',markersize=1)
axes[2].grid()
axes[2].set_ylim([-0.25, 6.25])
axes[2].set_ylabel("Pressure - Comp. \n(Bar)")

axes[3].plot(data[0]["t"], data[0]["v"],linestyle='None',marker='o',markersize=1)
axes[3].grid()
axes[3].set_ylabel("System (V)")

supply_frq = 2*np.pi/data[0]['t'].iloc[find_peaks(data[0]['p_o'])[0]].diff().mean()
center_val = data[0]['p_o'].mean()
amp = abs(data[0]['p_o'] - center_val).max()
```
